```python
from ultralytics import YOLO
import os

DATA_YAML_PATH = "C:/Python/strawberry-picker-cv/datasets/strawposesmall-1_1cls/data.yaml"
```

```python
model = YOLO("yolo26s-pose.pt")

results = model.train(
        data=DATA_YAML_PATH,
        epochs=100,                  # Количество эпох (проходов по датасету). Для начала 50-100 будет достаточно.
        imgsz=640,                  # К какому размеру приводить картинки при обучении
        batch=16,                   # Количество картинок в батче. Если выдает ошибку памяти GPU (OOM), снизьте до 8 или 4
        device=0,                   # 0 - использовать видеокарту (если установлен CUDA), 'cpu' - если только процессор
        project="strawberry_model", # Папка, куда сохранятся результаты обучения
        name="run_2",               # Имя текущего эксперимента
        plots=True                  # Генерировать графики обучения (появятся в папке с результатами)
    )
```

```python
metrics = model.val()
# Вывод некоторых метрик
print(f"Точность (mAP50) по рамкам: {metrics.box.map50:.3f}")
print(f"Точность (mAP50) по точкам: {metrics.pose.map50:.3f}")
```
