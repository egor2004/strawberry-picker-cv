# Преведение датасета2 от формата сегментации к формату ограничивающих рамок

```python
path_to_concat = path_to_dataset / "concat"
path_to_concat_images = path_to_concat / "images"
path_to_concat_labels_seg = path_to_concat / "labels"
```

```python
path_to_concat_labels_seg.rename(path_to_concat / "labels_seg")
```

```python
path_to_concat_labels = path_to_concat / "labels"
```

```python
path_to_concat_labels.mkdir()
```

```python
import cv2
import numpy as np
for item in path_to_concat_labels_seg.iterdir():
    mask = cv2.imread(str(item), cv2.IMREAD_UNCHANGED)
    height, width = mask.shape[:2]
    instance_ids = np.unique(mask)
    txt_path = path_to_concat_labels / item.with_suffix('.txt').name
    with txt_path.open('w') as txt_file:
        for inst_id in instance_ids:
            if inst_id == 0:
                continue # Пропускаем фон (0)

            # Находим координаты всех пикселей текущей клубники
            y_indices, x_indices = np.where(mask == inst_id)
            
            if len(x_indices) == 0 or len(y_indices) == 0:
                continue

            # Вычисляем пиксельные рамки
            x_min, x_max = np.min(x_indices), np.max(x_indices)
            y_min, y_max = np.min(y_indices), np.max(y_indices)

            bbox_width = (x_max - x_min) / width
            bbox_height = (y_max - y_min) / height
            center_x = (x_min + x_max) / 2.0 / width
            center_y = (y_min + y_max) / 2.0 / height

            # --- ДОБАВЛЕНИЕ КЛЮЧЕВЫХ ТОЧЕК ---
            v = 2 # Видимость точки (2 = размечена и видима)

            # Точка 1: Центр рамки (совпадает с center_x, center_y)
            kp1_x = center_x
            kp1_y = center_y

            # Точка 2: Верхняя точка рамки (середина верхней грани)
            # x совпадает с центром, y берем по самой верхней точке клубники (y_min)
            kp2_x = center_x
            kp2_y = y_min / height

            # Точка 3: На 50 пикселей выше второй точки
            kp3_x = center_x
            # Вычитаем 50 пикселей из y_min и нормализуем. Защищаем от ухода за край (max 0.0)
            kp3_y = max(0.0, (y_min - 50) / height)

            # Записываем строку в формате YOLO-Pose
            txt_file.write(
                f"{0} {center_x:.6f} {center_y:.6f} {bbox_width:.6f} {bbox_height:.6f} "
                f"{kp1_x:.6f} {kp1_y:.6f} {v} "
                f"{kp2_x:.6f} {kp2_y:.6f} {v} "
                f"{kp3_x:.6f} {kp3_y:.6f} {v}\n"
            )
```

```python
import cv2
import numpy as np
from pathlib import Path

# ВАЖНО: Добавьте путь к папке с оригинальными изображениями (откуда будем брать цвет)
# Предполагается, что файлы называются так же, как маски, но имеют расширение .jpg или .png
for item in path_to_concat_labels_seg.iterdir():
    # Читаем маску
    mask = cv2.imread(str(item), cv2.IMREAD_UNCHANGED)
    
    # Формируем путь к оригинальному изображению и читаем его
    # Если у вас изображения в формате .png, замените '.jpg' на '.png'
    img_path = path_to_concat_images / item.with_suffix('.png').name
    image = cv2.imread(str(img_path))
    
    # Проверка, что оригинальное изображение существует
    if image is None:
        print(f"Внимание: Не найдено изображение {img_path}. Пропуск.")
        continue

    height, width = mask.shape[:2]
    instance_ids = np.unique(mask)
    txt_path = path_to_concat_labels / item.with_suffix('.txt').name
    
    with txt_path.open('w') as txt_file:
        for inst_id in instance_ids:
            if inst_id == 0:
                continue # Пропускаем фон (0)

            # Находим координаты всех пикселей текущей клубники
            y_indices, x_indices = np.where(mask == inst_id)
            
            if len(x_indices) == 0 or len(y_indices) == 0:
                continue

            # Извлекаем все BGR пиксели конкретной ягоды из оригинальной картинки
            fruit_pixels = image[y_indices, x_indices]
            
            # В OpenCV порядок каналов: 0 - Blue, 1 - Green, 2 - Red
            # Вычисляем среднее значение зеленого и красного каналов
            mean_g = np.mean(fruit_pixels[:, 1])
            mean_r = np.mean(fruit_pixels[:, 2])
            
            # Логика: если красного в среднем больше, чем зеленого -> спелая (0)
            if mean_r > mean_g:
                class_id = 0
            else:
                class_id = 1  # иначе -> не спелая (1)
            # -----------------------------------

            # Вычисляем пиксельные рамки
            x_min, x_max = np.min(x_indices), np.max(x_indices)
            y_min, y_max = np.min(y_indices), np.max(y_indices)

            bbox_width = (x_max - x_min) / width
            bbox_height = (y_max - y_min) / height
            center_x = (x_min + x_max) / 2.0 / width
            center_y = (y_min + y_max) / 2.0 / height

            # --- ДОБАВЛЕНИЕ КЛЮЧЕВЫХ ТОЧЕК ---
            v = 2 # Видимость точки (2 = размечена и видима)

            # Точка 1: Центр рамки (совпадает с center_x, center_y)
            kp1_x = center_x
            kp1_y = center_y

            # Точка 2: Верхняя точка рамки (середина верхней грани)
            kp2_x = center_x
            kp2_y = y_min / height

            # Точка 3: На 50 пикселей выше второй точки
            kp3_x = center_x
            kp3_y = max(0.0, (y_min - 50) / height)

            # Записываем строку в формате YOLO-Pose (подставляем вычисленный class_id)
            txt_file.write(
                f"{class_id} {center_x:.6f} {center_y:.6f} {bbox_width:.6f} {bbox_height:.6f} "
                f"{kp1_x:.6f} {kp1_y:.6f} {v} "
                f"{kp2_x:.6f} {kp2_y:.6f} {v} "
                f"{kp3_x:.6f} {kp3_y:.6f} {v}\n"
            )
```
