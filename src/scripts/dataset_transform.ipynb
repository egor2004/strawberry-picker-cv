# Преобразование датасета1 к нужному формату папок

```python
from pathlib import Path

path_to_dataset = Path("../../datasets/strawberries/")
```

```python
path_to_train_img = path_to_dataset / "train/images"
path_to_train_lab = path_to_dataset / "train/lables"
path_to_train_img.mkdir(parents=True, exist_ok=True)
path_to_train_lab.mkdir(parents=True, exist_ok=True)
```

```python
path_to_training = path_to_dataset / "training"
for item in path_to_training.iterdir():
    if item.name[-4:] == ".jpg":
        item.rename(path_to_train_img / item.name)
    elif item.name[-4:] == ".txt":
        item.rename(path_to_train_lab / item.name)
```

```python
path_to_val_img = path_to_dataset / "val/images"
path_to_val_lab = path_to_dataset / "val/lables"
path_to_val_img.mkdir(parents=True, exist_ok=True)
path_to_val_lab.mkdir(parents=True, exist_ok=True)
```

```python
path_to_validation = path_to_dataset / "validation"
for item in path_to_validation.iterdir():
    if item.name[-4:] == ".jpg":
        item.rename(path_to_val_img / item.name)
    elif item.name[-4:] == ".txt":
        item.rename(path_to_val_lab / item.name)
```

```python
path_to_training.rmdir()
path_to_validation.rmdir()
```
